```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

```python
playercount = len(purchase_data["SN"].unique())
playercountframe = pd.DataFrame({"Total Players": [playercount]})
playercountframe
```

```python
unique_items = len(purchase_data["Item ID"].unique())
average_price = "${0:.3}".format(purchase_data["Price"].mean())
number_purchases = len(purchase_data["Item ID"])
total_revenue = (purchase_data["Price"].sum())
purchaseanalysisframe = pd.DataFrame({"Number of Unique Items" : [unique_items],
                                      "Average Price": [average_price],
                                      "Number of Purchases": [number_purchases],
                                      "Total Revenue": [total_revenue]
                                     })
purchaseanalysisframe
#format
```

```python
genderdemographics = purchase_data.groupby("Gender").apply(lambda x: len(x['SN'].unique()))
genderdemographics = pd.DataFrame(genderdemographics)
genderdemographics.columns = ["Total Count"]
totaluniqeplayers = len(purchase_data["SN"].unique())
genderdemographics["Percentage of Players"] = round((genderdemographics["Total Count"] /totaluniqeplayers *100),2)
genderdemographics
#format
```

```python

GenderAnalysis = purchase_data.groupby("Gender").agg({'Price': ["mean", "sum"]})
GenderAnalysis2 = purchase_data.groupby("Gender").apply(lambda x: len(x['SN']))
GenderAnalysis2 = pd.DataFrame(GenderAnalysis2)


uniquebygender = purchase_data.groupby("Gender").apply(lambda x: len(x['SN'].unique()))

GenderAnalysis= pd.merge(GenderAnalysis2, GenderAnalysis, on ="Gender")
GenderAnalysis.columns = ["Purchase Count", "Average Purchase Price", "Total Purchase Value"]
GenderAnalysis["Avg Total Purchase per Person"] = round((GenderAnalysis["Total Purchase Value"] / uniquebygender),2)
GenderAnalysis
#format
```

```python

bins = [0, 9, 14, 19, 24, 29, 34, 39, 200]
names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data2 = purchase_data
purchase_data2["Age Ranges"] = pd.cut(purchase_data2["Age"], bins, labels=names, include_lowest=True)


agedemographics = purchase_data2.groupby("Age Ranges").apply(lambda x: len(x['SN'].unique()))
agedemographics = pd.DataFrame(agedemographics)
agedemographics.columns = ["Total Count"]
agedemographics["Percentage of Players"] = round((agedemographics["Total Count"] /totaluniqeplayers *100),2)
agedemographics
#Format
```

```python


agedanalysis = purchase_data2.groupby("Age Ranges").agg({'Price': ['mean', "sum"]})
agedanalysis2 = purchase_data2.groupby("Age Ranges").apply(lambda x: len(x['SN']))
agedanalysis2 = pd.DataFrame(agedanalysis2)
agedanalysis = pd.merge(agedanalysis, agedanalysis2, on = "Age Ranges")

unqiuebyage = purchase_data2.groupby("Age Ranges").apply(lambda x: len(x['SN'].unique()))

agedanalysis.columns = ["Average Purchase Price", "Total Purchase Value", "Purchase Count"]
agedanalysis["Avg Total Purchase per Person"] = round((agedanalysis["Total Purchase Value"] / unqiuebyage),2)
agedanalysis

#format

```

```python

topspendertable = purchase_data.groupby("SN").agg({'SN': ['count'], 'Price': ["mean", "sum"]})
topspendertable.columns = ['Purchase Count', 'Average Purchase Price', 'Total Purchase Value']

topspendertable = topspendertable.sort_values(by=["Total Purchase Value"], ascending =0)
topspendertable.head(5)

#format
```

```python
mostpopitemstable = purchase_data.groupby(["Item ID", "Item Name"]).agg({'Item Name': ['count'], 'Price': ["sum"]})
mostpopitemstable.columns = ['Purchase Count', 'Total Purchase Value']
mostpopitemstable = pd.DataFrame(mostpopitemstable)

mostpopitemstable = mostpopitemstable.sort_values(by= ["Purchase Count"], ascending = 0)
mostpopitemstable["Item Price"] = mostpopitemstable['Total Purchase Value'] / mostpopitemstable['Purchase Count']

mostpopitemstable.head(5)
#format
```

```python
mostprotitemstable = purchase_data.groupby(["Item ID", "Item Name"]).agg({'Item Name': ['count'], 'Price': ["sum"]})
mostprotitemstable.columns = ['Purchase Count', 'Total Purchase Value']
mostprotitemstable = pd.DataFrame(mostprotitemstable)

mostprotitemstable = mostprotitemstable.sort_values(by= ["Total Purchase Value"], ascending = 0)

mostprotitemstable["Item Price"] = mostprotitemstable['Total Purchase Value'] / mostprotitemstable['Purchase Count']
mostprotitemstable.head(5)
#format
```
