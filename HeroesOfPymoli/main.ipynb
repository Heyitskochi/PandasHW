```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 200]
names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data2 = purchase_data
purchase_data2["Age Ranges"] = pd.cut(purchase_data2["Age"], bins, labels=names, include_lowest=True)
aged_table = purchase_data2.groupby("Age Ranges")
avgprice = aged_table["Price"].mean()
totalpurvalue = aged_table["Price"].sum()
temp1  = aged_table["Price"].sum()


agednum = list(aged_table["SN"].unique())
agedpricesum = list(aged_table["Price"].sum())
agednum = [len(num) for num in agednum]
avgpurperlist = []
for e in range(len(agedpricesum)):
    temp = agedpricesum[e] / agednum[e]
    avgpurperlist.append(temp)


finalagetable = pd.merge(avgprice, totalpurvalue, on = "Age Ranges")
finalagetable["Avg Total Purchase per Person"] = avgpurperlist
finalagetable
```
